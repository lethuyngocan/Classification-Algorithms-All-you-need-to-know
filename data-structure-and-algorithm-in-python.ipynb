# Sequence: String, List, Tuple:

### Index: can access any item in the sequence using index. Index starts with 0 for the first element.

```python
#string:
x= 'dragon'
print(x[3])

# list: 
x= ['cow', 'dragon', 'chicken', 'tiger','dog']
print(x[3])

#tuple: 

x= ('apple', 'pineapple','organge','strawberry')
print(x[3])
```

### Slicing: slice out substring, sublist, subtuple using index: 

```python
x= 'datascience'
print(x[1:6])
print(x[1:6:2])
print(x[3:])
print(x[:5])
print(x[:-2])
```

### Adding and concatenating: combine 2 sequence using +

```python
# string:
x= 'data' + 'science'

# list:
y = ['Machine']+['learning']
    
#tuple:
z=('py','thon','da','ta')+('scie','en')

print(x)
print(y)
print(z)
```

### Checking membership: check whether an item is in a not in a sequence:

```python
# string:
x='nature languge processing'

print ('u' in x)

# list:

y = ['cow', 'dragon', 'chicken', 'tiger','dog']
print('dog' in y)

# tuple: 
z= ('apple', 'pineapple','organge','strawberry')
print('melon' not in y)
```

### Minimum, maximum: find the minimum and maximum item in a sequence lexicographically:
### Alpha or numeric text but can not mix:

```python
# string:
x='nature languge processing'

print (min(x), max(x))

# list:

y = ['cow', 'dragon', 'chicken', 'tiger','dog']
print(min(y),max(y))

# tuple: 
z= ('apple', 'pineapple','organge','strawberry')
print(min(z),max(z))
```

### sum: find the sum of sequence in an item. Entire sequence must be numeric

```python
# list
y = [1,3,5,6,8,9]
print(sum(y))
print(sum(y[2:]))

# tuple:

z= (1,2,4,6,7,8,9,3)
print(sum(z))
print(sum(z[:-2]))
```

### Sort: returns a new list of items in sorted order:



```python
# string:
x= 'machinelearing'
print(sorted(x))

# list:
y = [1,3,5,6,8,9]
print(sorted(y))

# # tuple:
z= (1,2,4,6,7,8,9,3)
print(sorted(z))
```

### Count item: return count of an item


```python
#string:
x= 'hippo'
print(x.count('p'))

# list:
y = ['cow', 'dragon', 'chicken', 'tiger','dog']
print(y.count('cow'))

# tuple: 
z= ('apple', 'pineapple','organge','strawberry')
print(z.count('apple'))
```

### Index: return the index of the first occurence in an item: 


```python
#string:
x= 'hippo'
print(x.index('p'))

# list:
y = ['cow', 'dragon', 'chicken', 'tiger','dog']
print(y.index('cow'))

# tuple: 
z= ('apple', 'pineapple','organge','strawberry')
print(z.index('apple'))
```

### Unpacking: unpack the n items of a sequence into n variables

```python
z= ('apple', 'pineapple','organge','strawberry')
a,b,c,d =z
print(a,b,c,d)
```

# I.LIST

### List comprehensive:

```python
a = [m for m in range(8)]
print(a)

b = [i**2 for i in range(10) if i>4]
print(b)
```

### delete a list or an item in a list:

```python
x= [5,3,8,6]

del(x[1])
print(x)

```

### append- append an item to a list:



```python
x= [5,3,8,6]
x.append(7)
print(x)
```

### insert: inset an item at a given index

```python
x = [5,3,7,9]
x.insert(3,1) # (index, value)
print(x)
```

### pop: pop last item off list and returns item:

```python
x = [5,3,7,9]
x.pop()
print(x)
print(x.pop())
```

### remove: remove first instance of an item:



```python
x= [5,8,9,1,0,3,4,8]
x.remove(3)
print(x)
```

### reverse: reverse the order of the list. It is an in-place sort, meaning it changes the original list

```python
x= [5,8,9,1,0,3,4,8]
x.reverse()
print(x)
```

### sort: sort the list in place 
### sort(x) returns a new sorted list without changing the original list x
### x.sort() puts the items of x in sorted order (sort in place )

```python
x= [5,8,9,1,0,3,4,8]
x.sort()
print(x)
```

```python
x= [5,8,9,1,0,3,4,8]
sorted(x)
```

### reverse sort: sort items descending: use reverse = True 

```python
x= [5,8,9,1,0,3,4,8]
x.sort(reverse= True)
print(x)
```

### III. SETS
* store non-duplicate items
* very fast access vs list
* math set opp (union, intersect)
* set are Unodered



### Constructor: creating new sets


```python
x = { 3,5,8,9}
print(x)

y = set()
print(y)

list1= [2,4,6]
z= set(list1)
print(z)
```

### set operations:

```python
x = { 3,5,8,9}
print(x)
x.add(7)
print(x)
```

```python
x = { 3,5,8,9}
x.remove(3)
print(x)

#get the length of the set x:
print(len(x))

#check membership in x:
print(5 in x)

# pop random item from set x:
print(x.pop(),x)

# remove all the item from set x:
x.clear()
print(x)
```

### Mathematical set operations:
* interection (AND): set1&set2
* union(OR): set1|set2
* symmetric difference(XOR): set1^set2
* difference(in set1 but not set2): set1-set2
* subset(set2 contain set1) set1<=set2
* superset(set1 contain set2) set1>=set2

```python
s1= {1,2,4,7,8,4,8,4,8,1,7,9}
s2={4,7,9,4,8,9,3,7,0,8,4,}
print(s1&s2)
print(s1|s2)
print(s1^s2)
print(s2^s1)
print(s1-s2)
print(s1<=s2)
print(s1>=s2)
```
